## ATTRIBUTES OF AN OBJECT IN **PYTHON**

## 🔹 1. `.ndim`

Number of dimensions (axes) of the array

```python
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.ndim)
```

**Output:**

```
2
```

✔ Because it's a 2D array (rows & columns)

---

## 🔹 2. `.shape`

Size of each dimension (rows, columns, etc.)

```python
print(a.shape)
```

**Output:**

```
(2, 3)
```

✔ 2 rows, 3 columns

---

## 🔹 3. `.dtype`

Data type of elements in the array

```python
print(a.dtype)
```

**Output:**

```
int64
```
---

## 🔹 4. `.size`

Total number of elements

```python
print(a.size)
```

**Output:**

```
6
```

---

## 🔹 5. `.itemsize`

Size (in bytes) of each element

```python
print(a.itemsize)
```

---

## 🔹 6. `.nbytes`

Total bytes used by array

```python
print(a.nbytes)
```

---

## 🔹 7. `.T`

Transpose of the array

```python
print(a.T)
```

---

## 🔹 8. `.reshape()`

Change shape without changing data

```python
a.reshape(3, 2)
```

---

## 🔹 9. `.flatten()`

Convert to 1D array

```python
a.flatten()
```

---

## 🔹 10. `.astype()`

Convert data type

```python
a.astype(float)
```

# **TASK 2** - image flipping

````python
from PIL import Image
import matplotlib.pyplot as plt

image_2 = Image.open('car.jpg')
image_2_array = np.array(image_2)

plt.imshow(image_2_array)
plt.axis('off')
plt.show()
````

````python
horizontal = image_2_array[:, ::-1]
plt.imshow(horizontal)
plt.axis('off')
plt.show()
````

````python
vertical = image_2_array[::-1]
plt.imshow(vertical)
plt.axis('off')
plt.show()
````
