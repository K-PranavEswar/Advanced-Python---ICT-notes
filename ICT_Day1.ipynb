# **TASK 3** - SURVEY

````python
print("---------SURVEY 2026---------")
n = int(input("\nEnter number of students: "))

h = []

less_2 = 0
between_2_5 = 0
more_5 = 0

for i in range(1, n + 1):
    hours = float(input(f"Enter study hours for student {i}: "))
    h.append(hours)

    if hours < 2:
        less_2 += 1
    elif 2 <= hours <= 5:
        between_2_5 += 1
    else:
        more_5 += 1

average = sum(h) / len(h)

print("\n--- Survey Result ---")
print("Students studying < 2 hrs  :", less_2)
print("Students studying 2–5 hrs :", between_2_5)
print("Students studying > 5 hrs :", more_5)
print("Average study hours       :", round(average, 2))
````

# **What is a Function in Python?**

A **function** in Python is a block of reusable code that performs a specific task. Instead of writing the same code again and again, you can define a function once and call it whenever needed.

👉 **Definition:**
A function is a named set of statements that takes input, processes it, and returns an output.

---

## 🔹 Syntax of a Function

```python
def function_name(parameters):
    # block of code
    return value
```

---

## 🔹 Example of a Function

```python
def greet():
    print("Hello, Welcome to Python!")

greet()
```

---

## 🔹 Types of Functions in Python

Python functions are mainly classified into **two types**:

---

## 1️⃣ Built-in Functions

These are functions already provided by Python.

👉 Examples:


✔ No need to define them
✔ Directly usable

## 2️⃣ User-defined Functions

These are functions created by the user using `def`.

# **TASK 4 - SCORE CARD**

````python
students = {"Anu": (85, 90, 78),"Rahul": (60, 70, 65),"Meena": (95, 92, 88),"Kiran": (72, 68, 75)}
def total(marks): return sum(marks)
def average(marks): return sum(marks) / len(marks)
def grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "D"
print("----------- Student Score Card -----------")
print(f"{'Name':<10}{'Marks':<20}{'Total':<10}{'Average':<10}{'Grade':<5}")
print("-" * 55)
for name, marks in students.items():
    tot = total(marks)
    avg = average(marks)
    grd = grade(avg)
    print(f"{name:<10}{str(marks):<20}{tot:<10}{round(avg,2):<10}{grd:<5}")
````
