# **Lambda functions in python**

*finding a square of a number - normal method*

````python
def square(x):
    sq = x**2
    return sq
x = 2
square(x)
````

*instead we can apply lambda functions to reduce the code length,*

````python
square = lambda x:x**2
square(x)
````

````python
is_pass = lambda mark : "pass" if mark>60 else "fail"
is_pass(40)
````

*using map function*

````python
numbers = [10,20,30,40,15,25]
list(map(lambda x:x+5,numbers))
````

*using filter function*

````python
list(filter(lambda x:x>20,numbers))
````

````python
li = ['apple','banana','cherry','raspberry']
list(filter(lambda x: len(x) > 6, li))
````

*zip() is used to combine multiple lists (or iterables) element by element.*

````python
prices = [10,20,30,15]
quantities = [1,2,3,1]
item_prices = list(zip(prices, quantities))
item_prices
````

````python
prices = [10,20,30,15]
quantities = [1,2,3,1]
item_prices = list(zip(prices,quantities))
item_costs = list(map(lambda item : item[0]*item[1],item_prices))
item_costs
````

# **ERROR HANDLING & EXCEPTION HANDING**

# 🔹 Error Handling

**Error handling** is the process of managing errors in a program so it doesn’t crash.

👉 It ensures the program runs smoothly even when something goes wrong.

**Example:**

```python
a = 10
b = 0
print(a/b)   # causes error
```

---

# 🔹 Exception Handling

**Exception handling** is a method to handle runtime errors using special blocks like `try`, `except`.

👉 It prevents the program from stopping suddenly.

**Example:**

```python
try:
    a = 10
    b = 0
    print(a/b)
except:
    print("Error occurred")
```

# 🔹 `try` and `except`

### ✅ Definition:

* `try` → block where **error may occur**
* `except` → block that **handles the error**

# 🔹 Basic Syntax:

```python
try:
    # risky code
except:
    # runs if error occurs
```

# 🔹 Example:

```python
try:
    a = 10
    b = 0
    print(a / b)
except:
    print("Cannot divide by zero")
```

### 🔹 Output:

```

```

````python
class Student:
    def __init__(self, name, id, marks):
        self.name = name
        self.id = id
        self.marks = marks

    def average_score(self):
        return sum(self.marks) / len(self.marks)

    def pf(self):
        if self.average_score() >= 35:
            return "Passed"
        else:
            return "Failed"

    def display(self):
        print(f"-------------Student Details-------------")
        print(f"Name    : {self.name}")
        print(f"Rollno. : {self.id}")
        print(f"Marks   : {self.marks}")
        print(f"Average : {self.average_score()}")
        print(f"Pass/Fail : {self.pf()}")
Std1 = Student("Pranav",111,[60,77,88])
Std1.display()
Std2 = Student("Sooraj",112,[85,51,82])
Std2.display()
Std3 = Student("Adarsh",113,[77,70,90])
Std3.display()
Std4 = Student("Ambathoor chingam",114,[0,0,0])
Std4.display()
````

````python
class Student:
    def __init__(self, name, id, marks):
        self.name = name
        self.id = id
        self.marks = marks

    def total(self):
        return sum(self.marks)

    def average(self):
        return sum(self.marks) / len(self.marks)

    def grade(self):
        avg = self.average()
        if avg >= 90:
            return "A+"
        elif avg >= 75:
            return "A"
        elif avg >= 60:
            return "B"
        elif avg >= 50:
            return "C"
        else:
            return "F"

    def result(self):
        return "Pass" if self.average() >= 35 else "Fail"
students = [
    Student("Pranav", 111, [60, 77, 88]),
    Student("Sooraj", 112, [85, 51, 82]),
    Student("Adarsh", 113, [77, 70, 90])
]
students.sort(key=lambda s: s.total(), reverse=True)
print("="*90)
print(f"{'Rank':<6} {'Name':<10} {'RollNo':<10} {'Marks':<20} {'Total':<8} {'Avg':<8} {'Grade':<6} {'Result':<6}")
print("="*90)
for i, s in enumerate(students, start=1):
    print(f"{i:<6} {s.name:<10} {s.id:<10} {str(s.marks):<20} {s.total():<8} {s.average():<8.2f} {s.grade():<6} {s.result():<6}")
print("="*90)
````

# **INHERITANCE**

````python
class BankAccount:
    def __init__(self, holder, balance):
        self.holder = holder
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ₹{amount}. Current balance: ₹{self.balance}")
        else:
            print("Enter a valid amount")

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            print(f"₹{amount} withdrawn. Current balance: ₹{self.balance}")
        else:
            print("Insufficient balance or invalid amount")

    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, holder, balance, interest):
        super().__init__(holder, balance)
        self.interest = interest

    def apply_interest(self):
        interest_amount = self.balance * self.interest
        self.balance += interest_amount
        print(f"Interest added: ₹{interest_amount}")
        print(f"Updated balance: ₹{self.balance}")
customer1 = SavingsAccount("Alice", 5000, 0.05)
customer1.deposit(1000)
customer1.apply_interest()
customer1.withdraw(2000)
print("Final Balance:", customer1.get_balance())
````

````python
class BankAccount:
    def __init__(self, holder, balance):
        self.holder = holder
        self.balance = balance
        self.transactions = []   # passbook

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"Deposited ₹{amount}")
            print(f"Deposited ₹{amount}. Current balance: ₹{self.balance}")
        else:
            print("Enter a valid amount")

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            self.transactions.append(f"Withdrawn ₹{amount}")
            print(f"₹{amount} withdrawn. Current balance: ₹{self.balance}")
        else:
            print("Insufficient balance or invalid amount")

    def get_balance(self):
        return self.balance

    def add_interest(self):
        rate = 0.05   # 5% interest
        interest = self.balance * rate
        self.balance += interest
        self.transactions.append(f"Interest added ₹{interest:.2f}")
        print(f"Interest ₹{interest:.2f} added. New balance: ₹{self.balance:.2f}")

    def show_passbook(self):
        print("\n------ PASSBOOK ------")
        if not self.transactions:
            print("No transactions yet")
        else:
            for i, t in enumerate(self.transactions, start=1):
                print(f"{i}. {t}")
        print("----------------------")


# 🔹 User Input
name = input("Enter account holder name: ")
balance = float(input("Enter initial balance: "))

acc = BankAccount(name, balance)

while True:
    print("\n------ BANK MENU ------")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Add Interest")
    print("5. Passbook / Statement")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        amount = float(input("Enter amount to deposit: "))
        acc.deposit(amount)

    elif choice == "2":
        amount = float(input("Enter amount to withdraw: "))
        acc.withdraw(amount)

    elif choice == "3":
        print("Current Balance: ₹", acc.get_balance())

    elif choice == "4":
        acc.add_interest()

    elif choice == "5":
        acc.show_passbook()

    elif choice == "6":
        print("Thank you! 👋")
        break

    else:
        print("Invalid choice ❌")
````

# **FILE HANDLING**

**File handling** is the process of creating, reading, writing, updating, and deleting files using a programming language.

In Python, file handling allows you to **store data permanently** in files instead of keeping it only in memory.

---

## 🧠 Why File Handling?

* Store data permanently
* Read large data easily
* Save user inputs
* Work with logs, reports, databases

---

## 📁 Types of Files in Python

### 1️⃣ Text Files

* Store data in **readable format**
* Example: `.txt`, `.csv`, `.log`

📌 Example content:

```
Hello World
Python is easy
```

---

### 2️⃣ Binary Files

* Store data in **binary format (0s and 1s)**
* Not human-readable
* Example: `.jpg`, `.png`, `.mp3`, `.dat`

---

## 🔓 File Modes in Python

| Mode | Description         |
| ---- | ------------------- |
| `r`  | Read file           |
| `w`  | Write (overwrite)   |
| `a`  | Append              |
| `x`  | Create new file     |
| `b`  | Binary mode         |
| `t`  | Text mode (default) |
| `+`  | Read and write      |

---

## 📌 File Operations

---

### 1️⃣ Opening a File

```python
file = open("sample.txt", "r")
```

---

### 2️⃣ Reading a File

#### a) Read entire file

```python

```

#### b) Read one line

```python

```

#### c) Read all lines

```python

```

---

### 3️⃣ Writing to a File

#### a) Write (overwrite)

```python
file.write("Hello Python")
```

#### b) Append

```python

```

---

### 4️⃣ Closing a File

```python

```

---

Use **with statement** (auto closes file)

```python
with open("sample.txt", "r") as file:
    data = file.read()
```

---

## 🔄 File Handling Modes Explained

### 📖 Read Mode (`r`)

* Opens file for reading
* Error if file doesn't exist

---

### ✍️ Write Mode (`w`)

* Creates file if not exists
* Deletes old content

---

### ➕ Append Mode (`a`)

* Adds data at end
* Does not delete existing content

---

### 🆕 Create Mode (`x`)

* Creates new file
* Error if file exists

---

### 🔁 Read & Write (`r+`, `w+`, `a+`)

| Mode | Description              |
| ---- | ------------------------ |
| `r+` | Read + Write             |
| `w+` | Write + Read (overwrite) |
| `a+` | Append + Read            |

---
